```python
print("\n De acuerdo con la puntuación de felicidad del 2019" )
print("\n PAÍSES MÁS RICOS:") 
print(" La media obtenida fue:",df_masricos['Happiness Score 2019'].mean())
print(" El valor mínimo obtenido fue:",df_masricos['Happiness Score 2019'].min())
print(" El valor máximo obtenido fue:",df_masricos['Happiness Score 2019'].max())

print("\n PAÍSES MENOS RICOS:")
print(" La media obtenida fue:",df_menosricos['Happiness Score 2019'].mean())
print(" El valor mínimo obtenido fue:",df_menosricos['Happiness Score 2019'].min())
print(" El valor máximo obtenido fue:",df_menosricos['Happiness Score 2019'].max())
```

Revisando los resultados obtenidos en ambos años, podemos notar a simple vista que las puntuaciones de felicidad obtenidas en los países más ricos son más altas que en los países menos ricos.

## Gráficas multi-variantes.

```python
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x='Economy', y='Happiness Score 2015', hue='Clasification', 
           data=df_2015.loc[df_2015['Clasification'].isin([0, 1])], 
           fit_reg=False)
```

Notamos el mismo patrón de que los países desarrollados su felicidad está mayormente relacionada con el dinero.

```python
df_2015=df_2015.rename(columns = {'Economy (GDP per Capita) 2015': 'Economy'})


s = plt.hist(df_2015[df_2015["Clasification"] == 0].Economy,bins=30,label = "Subdesarrollado")
d = plt.hist(df_2015[df_2015["Clasification"] == 1].Economy,bins=30,label = "Desarrollado")
plt.legend()
plt.xlabel("Valor de GDP")
plt.ylabel("Frecuencia")
plt.title("Histograma")
plt.show()
```

Aquí observamos la frecuencia de como influye la felicidad en los países desarrollados y subdesarrollados.

```python
df_2015=df_2015.rename(columns = {'Generosity 2015': 'Generosity'})

mdata = pd.melt(df_2015,id_vars = "Clasification",value_vars = ['Economy', 'Generosity'])
plt.figure(figsize = (15,10))
sns.boxplot(x = "variable", y = "value", hue="Clasification",data= mdata)
plt.show()
```

En esta gráfica estamos comparando como influye la Economía y la Generosidad en la felicidad, ya que de acuerdo con lo que hemos visto, una influye más que la otra. Y lo que podemos observar es que efectivamente los valores de la economía son mucho más altos que el de la Generosidad.

```python
f,ax=plt.subplots(figsize = (18,18))
sns.heatmap(df_2015.corr(),annot= True,linewidths=0.5,fmt = ".1f",ax=ax)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Mapa de correlación')
plt.savefig('graph.png')
plt.show()
```

Aquí observamos la correlación de cada una de las variables pero nos enfocaremos por lo pronto en la correlación de la Economía y la Puntuación de la feliciad.
